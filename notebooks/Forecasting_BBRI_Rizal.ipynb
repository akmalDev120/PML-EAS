```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
data = yf.download("BBRI.JK", start="2020-01-01", end="2025-01-01")
data.head()
```

```python
data['Close'].plot(figsize=(12,5))
plt.title("Harga Penutupan Saham BBRI")
plt.show()
```

```python
from prophet import Prophet

df = data.reset_index()
df = df[['Date', 'Close']]
df.columns = ['ds', 'y']

model = Prophet()
model.fit(df)

future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

forecast[['ds','yhat']].tail()
```

```python
model.plot(forecast);
```

```python
forecast[['ds','yhat']].tail(10)
```

## Riset Model LSTM

### Pengertian
LSTM (Long Short-Term Memory) adalah salah satu jenis Recurrent Neural Network (RNN) yang dirancang untuk mempelajari pola pada data berurutan (time series).

### Kelebihan LSTM
- Mampu mengingat informasi jangka panjang.
- Cocok untuk prediksi harga saham.
- Dapat mempelajari pola data yang kompleks.

### Alur Kerja LSTM
Data Historis Saham
↓
Data Cleaning
↓
Normalisasi Data
↓
Pembuatan Sequence
↓
LSTM Layer
↓
Dense Layer
↓
Prediksi Harga Saham

### Perbandingan Prophet dan LSTM

| Prophet | LSTM |
|----------|----------|
| Mudah digunakan | Lebih kompleks |
| Cepat dilatih | Membutuhkan waktu training lebih lama |
| Cocok sebagai baseline | Cocok untuk model lanjutan |
| Tidak memerlukan banyak preprocessing | Memerlukan preprocessing data |
